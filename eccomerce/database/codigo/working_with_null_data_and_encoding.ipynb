##Codificando variáveis que não são numéricas

```python
encoder = LabelEncoder()

df["seller_id"] = encoder.fit_transform(df["seller_id"])
df["product_id"] = encoder.fit_transform(df["product_id"])
df["shipping_limit_date"] = encoder.fit_transform(df["shipping_limit_date"])
df["payment_type"] = encoder.fit_transform(df["payment_type"])
df["review_id"] = encoder.fit_transform(df["review_id"])
df["review_comment_title"] = encoder.fit_transform(df["review_comment_title"])
df["review_comment_message"] = encoder.fit_transform(df["review_comment_message"])
#df["review_creation_date"] = encoder.fit_transform(df["review_creation_date"])
df["review_answer_timestamp"] = encoder.fit_transform(df["review_answer_timestamp"])
df["customer_id"] = encoder.fit_transform(df["customer_id"])
df["order_status"] = encoder.fit_transform(df["order_status"])
df["order_purchase_timestamp"] = encoder.fit_transform(df["order_purchase_timestamp"])
#df["order_approved_at"] = encoder.fit_transform(df["order_approved_at"])
#df["order_delivered_carrier_date"] = encoder.fit_transform(df["order_delivered_carrier_date"])
df["order_delivered_customer_date"] = encoder.fit_transform(df["order_delivered_customer_date"])
#df["order_estimated_delivery_date"] = encoder.fit_transform(df["order_estimated_delivery_date"])
df["customer_unique_id"] = encoder.fit_transform(df["customer_unique_id"])
df["customer_city"] = encoder.fit_transform(df["customer_city"])
df["customer_state"] = encoder.fit_transform(df["customer_state"])
df["seller_city"] = encoder.fit_transform(df["seller_city"])
df["seller_state"] = encoder.fit_transform(df["seller_state"])
df["product_category_name"] = encoder.fit_transform(df["product_category_name"])

df.info()
```
